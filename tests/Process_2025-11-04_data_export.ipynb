```python
import numpy as np
from ham_reader import ham_reader
```

```python
raw_1Cu = ham_reader('../2025-11-04/10us_10meas_1Cu_30mJ.csv')
raw_2Al = ham_reader('../2025-11-04/10us_10meas_2Al_30mJ.csv')
raw_3Mg = ham_reader('../2025-11-04/10us_10meas_3Mg_30mJ.csv')
raw_4Zn = ham_reader('../2025-11-04/10us_10meas_4Zn_30mJ.csv')
raw_5Ti = ham_reader('../2025-11-04/10us_10meas_5Ti_30mJ.csv')
raw_6Fe = ham_reader('../2025-11-04/10us_10meas_6Fe_30mJ.csv')
raw_000 = ham_reader('../2025-11-04/10us_10meas_000_30mJ.csv')

raw_PLA = ham_reader('../2025-11-04/10us_1meas_PLA_50mJ.csv')
raw_PTFE = ham_reader('../2025-11-04/10us_1meas_PTFE_50mJ.csv')
raw_SUS = ham_reader('../2025-11-04/10us_10meas_SUS_30mJ.csv')

dark_1e0us = ham_reader('../2025-11-04/DARK_1e0us_1000meas.csv')
dark_1e1us = ham_reader('../2025-11-04/DARK_1e1us_1000meas.csv')
dark_1e2us = ham_reader('../2025-11-04/DARK_1e2us_1000meas.csv')
dark_1e3us = ham_reader('../2025-11-04/DARK_1e3us_1000meas.csv')
dark_1e4us = ham_reader('../2025-11-04/DARK_1e4us_100meas.csv')
dark_1e5us = ham_reader('../2025-11-04/DARK_1e5us_100meas.csv')
dark_1e6us = ham_reader('../2025-11-04/DARK_1e6us_100meas.csv')

raw_deuterium = ham_reader('../2025-11-04/LAMP_D_1e5us_100meas.csv')
raw_Hg_1e4us = ham_reader('../2025-11-04/LAMP_Hg_1e4us_100meas.csv')
raw_Hg_1e5us = ham_reader('../2025-11-04/LAMP_Hg_1e5us_100meas.csv')
raw_Hg_1e6us = ham_reader('../2025-11-04/LAMP_Hg_1e6us_10meas.csv')
```

```python
dark_1e1us_mean = np.mean(dark_1e1us, axis=0)

data_1Cu = raw_1Cu - dark_1e1us_mean
data_2Al = raw_2Al - dark_1e1us_mean
data_3Mg = raw_3Mg - dark_1e1us_mean
data_4Zn = raw_4Zn - dark_1e1us_mean
data_5Ti = raw_5Ti - dark_1e1us_mean
data_6Fe = raw_6Fe - dark_1e1us_mean
data_000 = raw_000 - dark_1e1us_mean
```

```python
px1 = 415
#px2 = 1016
wav1 = 213.60
#wav2 = 199.97
px2 = 1512
wav2 = 184.95

a = (wav2 - wav1) / (px2 - px1)
b = wav1 - a * px1

xdata = np.arange(0, len(data_1Cu[0]))

wav = a * xdata + b
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
```

```python
data_1 = [data_1Cu, data_2Al, data_3Mg, data_4Zn, data_5Ti, data_6Fe]
data_1_labels = ['Cu', 'Al', 'Mg', 'Zn', 'Ti', 'Fe']
label = '30 mJ, 10 us'

for i, data in enumerate(data_1):
    fig, ax = plt.subplots(figsize=(12, 3))
    ax.plot(wav, data[0])
    ax.set_xlim(180, 225)
    ax.set_title(f'{data_1_labels[i]}, {label}')
    ax.set_xticks(np.arange(180, 226, 2))
    ax.set_xlabel('Wavelength (nm)')
    ax.set_ylabel('Counts')
    ax.set_ylim(0,)
    fig.tight_layout()
    fig.savefig(f'../outputs/2025-11-04_testing_spectra/spectrum_LIBS_{data_1_labels[i]}_30mJ_10us.png', dpi=300)
    #plt.close(fig)
#plt.show()
```
